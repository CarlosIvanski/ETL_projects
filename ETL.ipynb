```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# ----- Dados -----
dados = {
    "Mês": ["jun/25", "jul/25", "ago/25", "set/25"],
    "Abertos": [34, 154, 278, 79],
    "Solucionados": [12, 74, 227, 162],
    "Fechado": [11, 76, 217, 151]
}

df = pd.DataFrame(dados)
x = np.arange(len(df["Mês"]))

# ----- Função para plotar série -----
def plot_montanha(x, y, label, cor):
    plt.plot(x, y, color=cor, linewidth=2, label=label, marker="o")
    plt.fill_between(x, y, color=cor, alpha=0.2)

# ----- Função para adicionar rótulos em cima ou embaixo -----
def adicionar_rotulos(x, y, indices_acima=None, indices_abaixo=None, cor="#000000"):
    if indices_acima:
        for i in indices_acima:
            plt.text(x[i], y[i] + 8, str(y[i]), ha="center", fontsize=11, color=cor, weight="bold")
    if indices_abaixo:
        for i in indices_abaixo:
            plt.text(x[i], y[i] - 12, str(y[i]), ha="center", fontsize=11, color=cor, weight="bold")

# ----- Gráfico maior -----
plt.figure(figsize=(12,8), facecolor="none")
ax = plt.gca()
ax.patch.set_alpha(0)

# --- Abertos ---
plot_montanha(x, df["Abertos"], "Abertos", "#4C9BE8")
adicionar_rotulos(
    x, df["Abertos"], 
    indices_acima=[0,2,3],   # jul/25, ago/25
    indices_abaixo=[1],  # jun/25, set/25
    cor="#4C9BE8"
)

# --- Solucionados ---
plot_montanha(x, df["Solucionados"], "Solucionados", "#AF710D")
adicionar_rotulos(
    x, df["Solucionados"], 
    indices_acima=[0,2,3],   # jul/25, ago/25
    indices_abaixo=[1],  # jun/25, set/25
    cor="#AF710D"
)

# --- Fechados ---
plot_montanha(x, df["Fechado"], "Fechado", "#07881C")
adicionar_rotulos(
    x, df["Fechado"], 
    indices_acima=[1],
    indices_abaixo=[0,2,3],
    cor="#07881C"
)


plt.xticks(x, df["Mês"], fontsize=11)
plt.yticks(fontsize=11)
plt.title("Evolução dos Chamados últimos 3 meses", fontsize=16)
plt.legend(fontsize=12)
plt.grid(alpha=0.25)

plt.tight_layout()
plt.show()
```

```python
# Pergunta 4:
# Escreva uma função, em qualquer linguagem de programação, que receba um número como parametro e retorne quantos numeros impares existem no
# intervalo de 0 até o numero recebebido.
# Exemplo:
# Parametro: 10
# Numeros impares no intervalo 0 até 10 = 5 -> (1, 3, 5, 7 e 9).

def validador_numero_impar(numero):
    if numero % 2 == 0:
        return numero / 2
    else:
        return ((numero - 1 ) / 2) + 1
    
print(validador_numero_impar(11))
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Dados
dados = {
    "Etapa": [
        "Leads Atendidos",
        "Lead Qualificado",
        "Colonoscopia (0)",
        "Endoscopia",
        "Exames",
        "Agendamento Humano (Especialidades)",
        "Agendado"
    ],
    "Quantidade": [543, 30, 0, 15, 18, 27, 89]
}

df = pd.DataFrame(dados)

cores = [
    "#00AEEF",   # Início
    "#F7941D",   # Lead Qualificado
    "#10C8B8",   # Colonoscopia
    "#7A7A7A",   # Endoscopia
    "#92278F",   # Exames
    "#8DC63F",   # Agendamento Humano
    "#F3008A"    # Agendado
]

# Plot
fig, ax = plt.subplots(figsize=(10,6), facecolor="none")

# Barras sem label 
bars = ax.bar(df.index, df["Quantidade"], color=cores, edgecolor="black")

# Rótulos de valor em cima das barras
for i, bar in enumerate(bars):
    height = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width()/2, height + 5,
        f"{height}", ha="center", va="bottom", fontsize=10
    )

# Título e eixo
ax.set_ylim(0, max(df["Quantidade"]) + 100)
ax.set_title("Relatório de Atendimentos - Período 22/09 → 01/10", fontsize=14)
ax.set_ylabel("Quantidade", rotation=0)

ax.set_xticks([])

# Estilo
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.yaxis.grid(True, linestyle="--", alpha=0.6)

# Legenda
handles = [plt.Rectangle((0,0),1,1, color=cor) for cor in cores]
ax.legend(
    handles, df["Etapa"], 
    loc="upper center", bbox_to_anchor=(0.5, -0.12),
    fancybox=True, shadow=False, ncol=4, fontsize=9,
    frameon=True, framealpha=1
)

ax.yaxis.grid(True, linestyle="--", alpha=0.6)
plt.grid(alpha=0.25)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dados
etapas = ["Lead Inicial", "Qualificação", "Conversão final"]
quantidade = [543, 30, 89]
total_inicial = quantidade[0]
cores = ['#66b3ff','#99ff99','#ffcc99']

# Legendas externas
legend_labels = [
    f"{etapas[i]} — {quantidade[i]} ({quantidade[i]/total_inicial*100:.1f}%)"
    for i in range(len(etapas))
]

# Função para mostrar apenas os valores absolutos na pizza
def autopct_absolute(pct, all_vals=quantidade):
    absolute = int(round(pct/100.*sum(all_vals)))
    return f"{absolute}"

# Pizza
fig, ax = plt.subplots(figsize=(8, 8), facecolor="none")
wedges, texts, autotexts = ax.pie(
    quantidade,
    labels=[None]*len(etapas),
    startangle=90,
    autopct=lambda pct: autopct_absolute(pct),
    wedgeprops=dict(edgecolor='black', linewidth=0.6),
    colors=cores,
    textprops=dict(color="black", fontsize=12, weight="bold")
)

# Legenda externa
ax.legend(
    wedges, 
    legend_labels, 
    title=None,
    loc='lower center',
    bbox_to_anchor=(0.5, -0.25),
    ncol=3,
    frameon=True,
    fancybox=False,
    edgecolor='black',
    fontsize=12
)

plt.tight_layout()
plt.show()
```
